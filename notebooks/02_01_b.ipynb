# Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10, 8
```

```python
rooms = 2*np.random.randn(100,1)+3
rooms[0:10]
```

```python
#creating a synthatic variable
price = 265 + 6*rooms + abs(np.random.randn(100,1))
price[0:10]
```

```python
plt.plot(rooms, price, 'r^')
plt.xlabel('# no of Rooms 2024 Average')
plt.ylabel('2024 Average Home Price $1000 USD')
plt.show()
```

```python
x = rooms
y = price

LinReg = LinearRegression()
LinReg.fit(x, y)
print(LinReg.intercept_, LinReg.coef_)
```

*Simple Algebra*
- y = mx + b
- b = intercept = 265.7

*Estimated Coefficients*
- LinReg.coef_ = [6.04] Estimated coefficients for the terms in the linear regression problem.

```python
#returns coefficient of determination which is the R squared of the prdiction
print(LinReg.score(x,y))
```

```python
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, LinReg.predict(x), color='red', linewidth=2, label='Regression line')
plt.xlabel('Number of Rooms')
plt.ylabel('Price')
plt.title('Linear Regression: Rooms vs Price')
plt.legend()
plt.grid(True)
```
